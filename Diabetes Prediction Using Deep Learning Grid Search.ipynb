```python
import sys
import pandas as pd
import numpy as np
import sklearn
import matplotlib.pyplot as plt

import keras
%matplotlib inline
```

```python
# import the uci diabetes dataset

url = "https://archive.ics.uci.edu/ml/machine-learning-databases/00529/diabetes_data_upload.csv"

df = pd.read_csv(url)
```

```python
df.describe(include = 'all')
```

```python
df.isnull().values.any()
```

```python
df['class'].value_counts()
```

```python
import seaborn as sns

bins = np.linspace(df.Age.min(), df.Age.max(), 10)
g = sns.FacetGrid(df, col="Gender", hue="class", palette="Set1", col_wrap=2)
g.map(plt.hist, 'Age', bins=bins, ec="k")

g.axes[-1].legend()
plt.show()
```

```python
g = sns.FacetGrid(df, col="Gender", hue="class", palette="Set1", col_wrap=2)
g.map(plt.bar, "Polyuria")

g.axes[-1].legend()
plt.show()
```
