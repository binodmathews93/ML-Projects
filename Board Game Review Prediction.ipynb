```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
```

```python
# Load the data

url = "https://github.com/ThaWeatherman/scrapers/raw/master/boardgamegeek/games.csv"
games = pandas.read_csv(url)
```

```python
# print the names of the columns in games

print(games.columns)

print(games.shape)
```

```python
# Plot a histogram of all the ratings in average_rating column

plt.hist(games["average_rating"])
plt.show()
```

```python
# print first row of games with average_rating as zero

print(games[games["average_rating"] == 0].iloc[0])

# print first row of games with average_rating greater than zero

print(games[games["average_rating"] > 0].iloc[0])
```

```python
# remove rows with no user ratings
games = games[games["users_rated"] > 0]

# remove rows with missing values
games.dropna(axis = 0)

# Plot a histogram of the average_rating column

plt.hist(games["average_rating"])
plt.show()
```

```python
# plot correlations between features

cormat = games.corr()
fig = plt.figure(figsize = (12,9))

sns.heatmap(cormat, vmax = 0.8, square = True)
plt.show
```
